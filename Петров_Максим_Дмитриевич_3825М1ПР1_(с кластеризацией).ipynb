```python
X = final_data.drop('HadHeartAttack', axis = 1)
y = final_data['HadHeartAttack']
```

Избавимся от коррелированных переменных, которые могут влиять на переобучение (Удаление коррелированных переменных - это как "уборка мусора" в данных, которая делает модели стабильнее, быстрее и надежнее!).
Возьмём корреляцию равную 0.8

```python

def remove_correlated_features_had_heart_attack(final_data, correlation_threshold=0.8):

    print(f"Исходное количество признаков: {X.shape[1]}")
    print(f"Размер данных: {final_data.shape}")

    corr_matrix = X.corr().abs()

    plt.figure(figsize=(14, 12))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=False, cmap='coolwarm', center=0,
                square=True, cbar_kws={"shrink": .8})
    plt.title('Матрица корреляций признаков')
    plt.tight_layout()
    plt.show()

    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    correlated_pairs = []
    for col in upper_tri.columns:
        for idx in upper_tri.index:
            if pd.notna(upper_tri.loc[idx, col]) and upper_tri.loc[idx, col] > correlation_threshold:
                correlated_pairs.append({
                    'feature1': idx,
                    'feature2': col,
                    'correlation': upper_tri.loc[idx, col]
                })

    to_drop = set()

    for pair in correlated_pairs:
        corr_with_target1 = abs(final_data[[pair['feature1'], 'HadHeartAttack']].corr().iloc[0,1])
        corr_with_target2 = abs(final_data[[pair['feature2'], 'HadHeartAttack']].corr().iloc[0,1])

        if corr_with_target1 >= corr_with_target2:
            to_drop.add(pair['feature2'])
            print(f"Сохраняем '{pair['feature1']}' (корр с целью: {corr_with_target1:.3f}), удаляем '{pair['feature2']}' (корр с целью: {corr_with_target2:.3f})")
        else:
            to_drop.add(pair['feature1'])
            print(f"Сохраняем '{pair['feature2']}' (корр с целью: {corr_with_target2:.3f}), удаляем '{pair['feature1']}' (корр с целью: {corr_with_target1:.3f})")

    X_reduced = X.drop(columns=to_drop)

    final_data_clean = pd.concat([X_reduced, y], axis=1)

    print(f"Найдено сильно коррелированных пар (correlation > {correlation_threshold}): {len(correlated_pairs)}")
    print(f"Удалено признаков: {len(to_drop)}")
    print(f"Исходное количество признаков: {X.shape[1]}")
    print(f"Итоговое количество признаков: {X_reduced.shape[1]}")

    if correlated_pairs:
        print(f"\nТоп-10 сильно коррелированных пар:")
        for i, pair in enumerate(correlated_pairs[:10]):
            print(f"  {i+1}. {pair['feature1']} - {pair['feature2']}: {pair['correlation']:.3f}")

    if to_drop:
        print(f"\nУдаленные признаки ({len(to_drop)}):")
        for i, feature in enumerate(sorted(to_drop)):
            print(f"  {i+1}. {feature}")

    if len(to_drop) > 0:
        plt.figure(figsize=(12, 10))
        corr_after = X_reduced.corr().abs()
        mask_after = np.triu(np.ones_like(corr_after, dtype=bool))
        sns.heatmap(corr_after, mask=mask_after, annot=False, cmap='coolwarm', center=0,
                   square=True, cbar_kws={"shrink": .8})
        plt.title('Матрица корреляций ПОСЛЕ удаления')
        plt.tight_layout()
        plt.show()

    return final_data_clean, to_drop, correlated_pairs

final_data_clean, dropped_features, correlated_pairs = remove_correlated_features_had_heart_attack(
    final_data,
    correlation_threshold=0.8
)

X = final_data_clean.drop('HadHeartAttack', axis=1)
y = final_data_clean['HadHeartAttack']

print(f"X shape: {X.shape}")
print(f"y value counts:")
print(y.value_counts())
```

Разобьем данные на обучающую и тестовую выборки в пропорции 3:1 (75% - обучающая выборка, 25% - тестовая):

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42, stratify=y)
```

Для сравнения алгоритмов будем использовать Recall, потому что часто лучше здорового назвать больным, чем пропустить больного.

# Для начала возьмём один из простых методов - KNN

Попробуем подобрать оптимальные гиперпараметры

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix, recall_score
```

```python
from sklearn.model_selection import StratifiedKFold

stratified_cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

sample_size = 100000  # подвыборка для скорости

# Используем train_test_split для стратификации
X_sample, _, y_sample, _ = train_test_split(
    X_train, y_train,
    train_size=min(sample_size, len(X_train)),
    stratify=y_train,  # сохраняем пропорции классов
    random_state=42
)

print(f"Размер подвыборки: {len(X_sample)}")
print(f"Класс 0: {(y_sample == 0).sum()} ({np.mean(y_sample == 0):.1%})")
print(f"Класс 1: {(y_sample == 1).sum()} ({np.mean(y_sample == 1):.1%})")
```

```python
#  Параметры для поиска
param_grid = {
    'n_neighbors': [5, 7, 9, 15],
    'weights': ['distance'],
    'p': [2]
}

grid_search = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,  # параметры для перебора
    cv=stratified_cv,
    scoring='recall',
    n_jobs=-1
)

print("\nНачинаем подбор параметров...")
grid_search.fit(X_sample, y_sample)  # обучаем на подвыборке

# Выводим результаты
print(f"\nЛучшие параметры: {grid_search.best_params_}")
print(f"Лучший recall на кросс-валидации: {grid_search.best_score_:.4f}")
```

```python
best_params_knn = grid_search.best_params_
final_knn = KNeighborsClassifier(**grid_search.best_params_)
final_knn.fit(X_train, y_train)

# Проверяем на тестовых данных
y_pred = final_knn.predict(X_test)
```

```python
dbscan = DBSCAN(eps=2.0, min_samples=10, n_jobs=-1) # при eps > 2 нехватка ОЗУ
dbscan_labels = dbscan.fit_predict(X_train)
```

```python
kmeans_labels_pca = kmeans_labels
dbscan_labels_pca = dbscan_labels
gmm_labels_pca = gmm_labels
```

ARI (Adjusted Rand Index) - это метрика для оценки качества кластеризации, которая сравнивает полученные кластеры с истинными метками

```python
if -1 in dbscan_labels:
    non_noise = dbscan_labels != -1
    noise_count = np.sum(~non_noise)
    valid_count = np.sum(non_noise)

    print(f"\nDBSCAN статистика:")
    print(f"  Шумовых точек: {noise_count} ({noise_count/len(dbscan_labels)*100:.1f}%)")
    print(f"  Валидных точек: {valid_count}")

    if valid_count > 0:
        ari_dbscan = adjusted_rand_score(y_train[non_noise], dbscan_labels[non_noise])
        print(f"  DBSCAN ARI (без шума): {ari_dbscan:.4f}")
    else:
        print("  ВНИМАНИЕ: DBSCAN определил все точки как шум!")
else:
    ari_dbscan = adjusted_rand_score(y_train, dbscan_labels)
    print(f"DBSCAN ARI: {ari_dbscan:.4f} (шума нет)")
```
